```python
# converting the release date into date timeformat
df['Release_Date']=pd.to_datetime(df['Release_Date'])
print(df['Release_Date'].dtype)
```

```python
# covnerting into the year
df['Release_Date']=df['Release_Date'].dt.year
df['Release_Date'].dtypes
```

```python
# checking
df.head()
```

```python
# removing the columns which are not needed
cols=['Overview','Original_Language','Poster_Url']
df.drop(cols,axis=1,inplace=True)
```

```python
df.head()
```

```python
# creating the label's fo reach area like popular, avearage,below average & not pupular
# creating th ecoloumn name categorys
# Creating the dunction coloumn
def catigroize_col(df,col,labels):
    edges = [
        df[col].describe()['min'],
        df[col].describe()['25%'],
        df[col].describe()['50%'],
        df[col].describe()['75%'],
        df[col].describe()['max']]
    df[col]=pd.cut(df[col],edges,labels=labels,duplicates='drop')
    return df
               
```

```python
labelz=['most_popular','below_avg','average','popular']
catigroize_col(df,'Vote_Average',labelz)
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
# Breaking the genre into new coloumn
#example :
# SpiderMan :Action
# SpiderMan :Adventure
# SpiderMan :Science
# SpiderMan :Fiction

df['Genre']=df['Genre'].str.split(', ')
df=df.explode('Genre').reset_index(drop=True)
df.head()
```
